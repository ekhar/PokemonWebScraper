```python
from bs4 import BeautifulSoup as soup
import urllib
import pandas as pd
import os
```

```python
folder = "/home/eric/Desktop/Pokemon_DataScience/pokemon_images"
url = "https://pokemondb.net/pokedex/national"
```

```python
req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
page = urllib.request.urlopen(req)
page_html =  page.read()
page.close()
```

```python
page_soup = soup(page_html, "html.parser")
```

```python
cards = page_soup.findAll("div", {"class":"infocard"})
```

```python
#Objective: get list of variables image_link, number,pokemon_name, 
#type1, and if applicable type 2

numbers_list = []
names_list = []
type1_list = []
type2_list = []
png_list = []


for card in cards:    
    
    #Goes straight to the important html text and has it in a list
    important_info = card.select("span.infocard-lg-data")[0].find_all(True)
    
    image_link = card.a.span["data-src"] #image link
    number = int(important_info[0].text[1:])
    pokemon_name = important_info[2].text
    type1 = important_info[5].text
    #if the pokemon has a Type2, it will be in the 6th index
    try:
        type2 = important_info[6].text
    except:
        type2 = None

    
    numbers_list.append(number)
    names_list.append(pokemon_name)
    type1_list.append(type1)
    type2_list.append(type2)
    png_list.append(image_link)
```

```python
pokedata_list = [numbers_list,names_list,type1_list,type2_list,png_list]
```

```python
df = pd.DataFrame(pokedata_list)
```

```python
df = df.transpose()
```

```python
df.columns = ["Pokedex Number", "Name", "Type1", "Type2", "PNG Link"]
```

```python
folder = "/home/eric/Desktop/Pokemon_DataScience/pokemon_images/"

#names the pokemon according to their pokedex entry
for count, link in enumerate(df['PNG Link']):
    fname = str(count+1)
    filename = folder + fname
    
    image_url = df['PNG Link'][count]
    
    #downloads the images
    opener=urllib.request.build_opener()
    opener.addheaders=[('User-Agent','Mozilla/5.0')]
    urllib.request.install_opener(opener)
    
    urllib.request.urlretrieve(image_url,filename)
```
